```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
```

```python
data = pd.read_csv('EnglandWeather.csv')
df = pd.DataFrame(data)
```

# PreProcessing

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc = True).dt.year
```

```python
df['Summary'].unique()
```

```python
df['Precip Type'].unique()
```

```python
df = df.dropna()
```

```python
df = df[df['Wind Speed (km/h)'] > 0]
df = df[df['Humidity'] > 0]
df = df[df['Pressure (millibars)'] > 0]
```

```python
df = df.reset_index()
```

```python
le = LabelEncoder()
# change Summary column numerically
le.fit(df['Summary'])
df['Summary'] = le.transform(df['Summary'])
```

```python
# change Precip Type column numerically
le.fit(df['Precip Type'])
df['Precip Type'] = le.transform(df['Precip Type'])
```

```python
df
```

```python
df = df.drop('index', axis=1)
```

```python
df
```

# Charts

```python
plt.title("Humidity distribution")
x = df['Humidity']
plt.hist(x)
```

```python
y = df['Humidity']
x = df['Temperature (C)']
plt.scatter(x, y)
```

```python
y = df['Humidity']
x = df['Pressure (millibars)']
plt.scatter(x, y)
```

```python
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(x, y)
def my_func(x):
  return slope * x + intercept

my_model = list(map(my_func, x))
plt.xlabel('Pressure (millibars)')
plt.ylabel('Humidity')
plt.scatter(x, y, color='green')
plt.plot(x, my_model, color='red')
```
