```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from kneed import KneeLocator
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
import seaborn as sns
```

```python
data = pd.read_csv('Customer_Data.csv')
df = pd.DataFrame(data)
```

# Preprocessing

```python
df = df.drop('CUST_ID', axis=1)
```

```python
df[df['MINIMUM_PAYMENTS'].isnull() == True] = int(df['MINIMUM_PAYMENTS'].mean())
df[df['CREDIT_LIMIT'].isnull() == True] = int(df['CREDIT_LIMIT'].mean())
```

```python
print(df.isnull().sum())
```

```python
sns.histplot(df['BALANCE'])
plt.show()
```

```python
for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    mi = Q1 - 1.5 * IQR
    ma = Q1 + 1.5 * IQR

    df[df[col] < mi] = int(df[col].mean())
    df[df[col] > ma] = int(df[col].mean())
```

```python
Q1 = df['BALANCE'].quantile(0.25)
Q3 = df['BALANCE'].quantile(0.75)
IQR = Q3 - Q1
mi = Q1 - 1.5 * IQR
ma = Q1 + 1.5 * IQR
```

```python
df[df['BALANCE'] < mi] = int(df['BALANCE'].mean())
df[df['BALANCE'] > ma] = int(df['BALANCE'].mean())
```

```python
print(df.describe())
```

```python
sns.histplot(df['BALANCE'])
plt.show()
```

```python
cols = df.columns
scaler = StandardScaler()
scaled_col = scaler.fit_transform(df)
df = pd.DataFrame(scaled_col, columns=[cols])
```

```python
df
```

# KMeans

```python
km = KMeans(n_clusters=15, random_state=40, n_init=10)
km.fit_predict(df)
```

```python
score = silhouette_score(df, km.labels_, metric='euclidean')
print('Silhouetter Average Score: %.3f' % score)
```

# Agglomerative Clustering

```python
model = AgglomerativeClustering(n_clusters=12, linkage='ward')
labels = model.fit_predict(df)
```

```python
silhouette_avg = silhouette_score(df, labels)
print("Silhouette Score:", silhouette_avg)
```

```python
kmeans_set={"init":"random","n_init":10,"max_iter":300,"random_state":42}
cluster_range = range( 2, 10 )
cluster_errors=[]
for k in cluster_range:
    kmeans= KMeans(n_clusters=k, **kmeans_set)
    kmeans.fit(df)
    score = silhouette_score(df, kmeans.labels_, metric='euclidean')
    print('Silhouetter Average Score: %.3f' % score, 'K = ', k)
    cluster_errors.append( kmeans.inertia_) 
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2,10),cluster_errors)
plt.xticks(range(2,10))
plt.title('Elbow Method For Optimal k')
plt.xlabel('number of clusters')
plt.ylabel('inertia')
plt.show()
```

```python
k1=KneeLocator(range(2,10),cluster_errors , curve='convex', direction= 'decreasing')
k1.elbow
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2,10),cluster_errors)
plt.xticks(range(2,10))
plt.xlabel('number of clusters')
plt.ylabel('List')
plt.title('Elbow Method For Optimal k')
plt.axvline(x=k1.elbow, color='b', label= 'axvline-full height', ls= '--')
plt.show()
```

# Calinski Harabaz

```python
calinski_harabaz = []
for n_clust in range(2, 11):
    AgglomerativeC = AgglomerativeClustering(n_clusters=n_clust, metric='euclidean', linkage='ward')
    AgglomerativeC.fit(df)
    score = metrics.calinski_harabasz_score(df, AgglomerativeC.labels_)
    calinski_harabaz.append(score)
with plt.style.context("fivethirtyeight"):
    fig, ax = plt.subplots()
    ax.plot(range(2, 11), calinski_harabaz, marker='^', c='b', ms=9, mfc='r')
    ax.set_xticks(range(2, 11))
    ax.set_xlabel("Number of Clusters")
    ax.set_ylabel("Calinski-Harabasz Score")
    ax.set_title("Calinski-Harabasz Plot")
    plt.show()
```

```python
calinski_harabaz_scores = []
for k in range(2, 11):
    AgglomerativeC = AgglomerativeClustering(n_clusters=k, linkage='ward')
    labels = AgglomerativeC.fit_predict(df)
    calinski_harabaz_score_k = calinski_harabasz_score(df, labels)
    calinski_harabaz_scores.append(calinski_harabaz_score_k)
scores_Agg = pd.DataFrame({'k': range(2,11), 'Calinski-Harabasz Score': calinski_harabaz_scores})
print("\nScores for Different Numbers of Clusters:")
print(scores_Agg)
best_k_calinski_harabaz = scores_Agg.loc[scores_Agg['Calinski-Harabasz Score'].idxmax(), 'k']
print(f"Best k for Calinski-Harabasz Score: {best_k_calinski_harabaz}")
```
