```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('heart_statlog_cleveland_hungary_final.csv')
df = pd.DataFrame(data)
```

```python
from sklearn.model_selection import train_test_split

x = df.drop('target', axis=1)
y = df['target']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(max_depth=15, min_samples_split=18, min_samples_leaf=3)
dt.fit(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=103, max_depth=18)
rf.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred_train_dt = dt.predict(x_train)
y_pred_test_dt = dt.predict(x_test)

y_pred_train_rf = rf.predict(x_train)
y_pred_test_rf = rf.predict(x_test)

acc_train_dt = accuracy_score(y_train, y_pred_train_dt)
acc_test_dt = accuracy_score(y_test, y_pred_test_dt)

acc_train_rf = accuracy_score(y_train, y_pred_train_rf)
acc_test_rf = accuracy_score(y_test, y_pred_test_rf)

acc_train_dt, acc_test_dt, acc_train_rf, acc_test_rf
```
